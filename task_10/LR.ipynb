```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import pylab as pl
%matplotlib inline

def pz(fico, amt, coef):
    z = coef[0]*fico + coef[1]*amt + coef[2]
    return 1 / (1 + np.exp(-1 * z))

def main():
    loanData = pd.read_csv('loansData.csv')
    loanData = loanData.dropna(how = 'any')
    loanData = loanData[['Interest.Rate', 'FICO.Range', 'Loan.Length', 'Monthly.Income', 'Amount.Requested']]
    print loanData.head()
    
    ##Избавляемся от %
    f = loanData['Interest.Rate']
    f = [float(a[:-1]) for a in f]
    loanData['Interest.Rate'] = f    
    loanData['TF'] = loanData['Interest.Rate']<=14
    
    ##Избавляемся от второго значения в FICO
    f = loanData['FICO.Range']
    f = [int(a[:3]) + 2 for a in f]
    loanData['FICO.Score'] = f
    
    ##Выбирем изначально константу 1
    loanData['intercept'] = 1.0
    ind_cols=['FICO.Score','Amount.Requested','intercept']
    logit = sm.Logit(loanData['TF'], loanData[ind_cols])
    result = logit.fit()
    coeff = result.params
    
    xlist = range(0, 60000, 10)
    ylist = [pz(820, x, coeff) for x in xlist]
    pl.plot(xlist, ylist)
    pl.show()
    
    xlist = range(0, 60000, 10)
    ylist = [pz(720, x, coeff) for x in xlist]
    pl.plot(xlist, ylist)
    pl.show()

if __name__ == '__main__':
    main()
```
